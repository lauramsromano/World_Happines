```python
#Importando biblioteca necessária para tratamento dos dados:

import pandas as pd
from datetime import datetime
```

```python
#Lista com os nomes das colunas importantes em comum entre os dados dos diferentes anos:
common_cols = ['Happiness Rank', 'Country', 'Region', 'Happiness Score', 'Economy (GDP per Capita)', 'Family (Social Support)','Health (Life Expectancy)','Freedom', 'Trust (Government Corruption)','Generosity','Year']
```

```python
#Importando as tabelas:

dados_2015 = pd.read_csv("2015.csv")
dados_2016 = pd.read_csv("2016.csv")
dados_2017 = pd.read_csv("2017.csv")
dados_2018 = pd.read_csv("2018.csv")
dados_2019 = pd.read_csv("2019.csv")
dados_2020 = pd.read_csv("2020.csv")
dados_2021 = pd.read_csv("2021.csv")
dados_2022 = pd.read_csv("2022.csv")

#Convertendos os anos para tipo 'data':
ano_2015 = datetime.strptime('2015', '%Y').date()
ano_2016 = datetime.strptime('2016', '%Y').date()
ano_2017 = datetime.strptime('2017', '%Y').date()
ano_2018 = datetime.strptime('2018', '%Y').date()
ano_2019 = datetime.strptime('2019', '%Y').date()
ano_2020 = datetime.strptime('2020', '%Y').date()
ano_2021 = datetime.strptime('2021', '%Y').date()
ano_2022 = datetime.strptime('2020', '%Y').date()

#Adicionando coluna de Year (ano) nos dataframes:
years = [ano_2015, ano_2016, ano_2017, ano_2018, ano_2019, ano_2020, ano_2021, ano_2022]
df = [dados_2015, dados_2016, dados_2017, dados_2018, dados_2019, dados_2020, dados_2021, dados_2022]

for i in range(0,8):
    df[i]['Year'] = years[i]
```

```python
#Renomeando colunas necessárias:
dados_2015 = dados_2015.rename(columns={'Family':'Family (Social Support)'})

dados_2015.info()
```

```python
#Renomeando colunas necessárias:
dados_2016 = dados_2016.rename(columns={'Family':'Family (Social Support)'})

dados_2016.info()
```

```python
#Renomeando colunas necessárias:
dados_2017 = dados_2017.rename(columns = {"Whisker.high": "Whisker high",
                                          "Whisker.low": "Whisker low",
                                          "Happiness.Rank": "Happiness Rank",
                                          "Happiness.Score": "Happiness Score",
                                          "Economy..GDP.per.Capita.":"Economy (GDP per Capita)",
                                          "Health..Life.Expectancy.":"Health (Life Expectancy)",
                                          "Trust..Government.Corruption.":"Trust (Government Corruption)",
                                          "Dystopia.Residual":"Dystopia Residual",
                                          'Family':'Family (Social Support)',})


#Criando e atribuindo valores a coluna Região a parte da coluna Região já existente no df com dados de 2015:
dados_2017 = dados_2017.merge(dados_2015[["Country","Region"]], on="Country", how="left")
#Colocando '-' onde não bá região correspondente para tratamentos mais a frente:
dados_2017["Region"] = dados_2017["Region"].fillna('-')

dados_2017.info()
```

```python
#Renomeando colunas necessárias:
dados_2018 = dados_2018.rename(columns = {"Overall rank": "Happiness Rank",
                                          "Score": "Happiness Score",
                                          "Country or region": "Country",
                                          "Social support":"Family (Social Support)",
                                          "GDP per capita":"Economy (GDP per Capita)",
                                          "Healthy life expectancy":"Health (Life Expectancy)",
                                          "Freedom to make life choices":"Freedom",
                                          "Perceptions of corruption":"Trust (Government Corruption)"})

#Criando e atribuindo valores a coluna Região a parte da coluna Região já existente no df com dados de 2015:
dados_2018 = dados_2018.merge(dados_2015[["Country","Region"]], on="Country", how="left")
#Colocando '-' onde não bá região correspondente para tratamentos mais a frente:
dados_2018["Region"] = dados_2018["Region"].fillna('-')

dados_2018.info()
```

```python
#Renomeando colunas necessárias:
dados_2019 = dados_2019.rename(columns = {"Overall rank": "Happiness Rank", 
                             "Country or region": "Country", 
                             "GDP per capita":"Economy (GDP per Capita)",
                             "Healthy life expectancy":"Health (Life Expectancy)",
                             "Freedom to make life choices":"Freedom",
                             "Score" : "Happiness Score", 
                             "Social support":"Family (Social Support)",
                             "Perceptions of corruption":"Trust (Government Corruption)"})

#Criando e atribuindo valores a coluna Região a parte da coluna Região já existente no df com dados de 2015:
dados_2019 = dados_2019.merge(dados_2015[["Country","Region"]], on="Country", how="left")
#Colocando '-' onde não bá região correspondente para tratamentos mais a frente:
dados_2019["Region"] = dados_2019["Region"].fillna('-')

dados_2019.info()
```

```python
#Removendo colunas desnecessárias:
dados_2020.drop('Standard error of ladder score', axis = 1, inplace=True)
dados_2020.drop('Logged GDP per capita', axis = 1, inplace=True)
dados_2020.drop('Social support', axis = 1, inplace=True)
dados_2020.drop('Healthy life expectancy', axis = 1, inplace=True)
dados_2020.drop('Freedom to make life choices', axis = 1, inplace=True)
dados_2020.drop('Generosity', axis = 1, inplace=True)
dados_2020.drop('Perceptions of corruption', axis = 1, inplace=True)
dados_2020.drop('Ladder score in Dystopia', axis = 1, inplace=True)

#Renomeando colunas necessárias:
dados_2020 = dados_2020.rename(columns={'Country name':'Country',
                                'Regional indicator': 'Region',
                                "upperwhisker": "Whisker high",
                                "lowerwhisker": "Whisker low",
                                'Ladder score':'Happiness Score',
                                "Explained by: Social support":"Family (Social Support)",
                                "Explained by: Healthy life expectancy":"Health (Life Expectancy)",
                                "Explained by: Freedom to make life choices":"Freedom",
                                "Explained by: Perceptions of corruption":"Trust (Government Corruption)",
                                "Explained by: Log GDP per capita":"Economy (GDP per Capita)",
                                "Explained by: Generosity":"Generosity",
                                "Dystopia + residual" :"Dystopia"})

#Criando e atribuindo valores a coluna Happiness Rank utilizando um range considerando o index das linhas:
dados_2020['Happiness Rank'] = [i for i in range(1, len(dados_2020.index)+1)]


dados_2020.info()
```

```python
#Removendo colunas desnecessárias:
dados_2021.drop('Standard error of ladder score', axis = 1, inplace=True)
dados_2021.drop('Logged GDP per capita', axis = 1, inplace=True)
dados_2021.drop('Social support', axis = 1, inplace=True)
dados_2021.drop('Healthy life expectancy', axis = 1, inplace=True)
dados_2021.drop('Freedom to make life choices', axis = 1, inplace=True)
dados_2021.drop('Generosity', axis = 1, inplace=True)
dados_2021.drop('Perceptions of corruption', axis = 1, inplace=True)
dados_2021.drop('Ladder score in Dystopia', axis = 1, inplace=True)

#Renomeando colunas necessárias:
dados_2021 = dados_2021.rename(columns={'Country name':'Country',
                                'Regional indicator': 'Region',
                                "upperwhisker": "Whisker high",
                                "lowerwhisker": "Whisker low",
                                'Ladder score':'Happiness Score',
                                'Explained by: Social support':'Family (Social Support)',
                                'Explained by: Healthy life expectancy':'Health (Life Expectancy)',
                                'Explained by: Freedom to make life choices':'Freedom',
                                'Explained by: Perceptions of corruption':'Trust (Government Corruption)',
                                'Explained by: Log GDP per capita':'Economy (GDP per Capita)',
                                'Explained by: Generosity':'Generosity',
                                "Dystopia + residual" :"Dystopia"})


#Criando e atribuindo valores a coluna Happiness Rank utilizando um range considerando o index das linhas:
dados_2021['Happiness Rank'] = [i for i in range(1, len(dados_2021.index)+1)]

dados_2021.info()
```

```python
#Renomeando colunas necessárias:
dados_2022 = dados_2022.rename(columns={'RANK':'Happiness Rank',
                                'Happiness score':'Happiness Score',
                                'Explained by: GDP per capita':'Economy (GDP per Capita)',
                                'Explained by: Social support':'Family (Social Support)',
                                'Explained by: Healthy life expectancy':'Health (Life Expectancy)',
                                'Explained by: Freedom to make life choices':'Freedom',
                                'Explained by: Generosity':'Generosity',
                                'Explained by: Perceptions of corruption':'Trust (Government Corruption)'})

#Criando e atribuindo valores a coluna Região a parte da coluna Região já existente no df com dados de 2015:
dados_2022 = dados_2022.merge(dados_2015[["Country","Region"]], on="Country", how="left")
#Colocando '-' onde não bá região correspondente para tratamentos mais a frente:
dados_2022["Region"] = dados_2022["Region"].fillna('-')

dados_2022.info()
```

```python
#Lista com as colunas em comum nos dataframes
dados_filtrados = [dados_2015[common_cols], dados_2016[common_cols], dados_2017[common_cols], dados_2018[common_cols], dados_2019[common_cols], dados_2020[common_cols], dados_2021[common_cols], dados_2022[common_cols]]
```

```python
#Criando o dataframe final
dados_finais = pd.DataFrame(columns=common_cols)
dados_finais = dados_finais.append(dados_filtrados)

#Removendo linhas com valores nulos:
dados_finais.dropna(axis='rows',inplace=True)

dados_finais
```

```python
dados_finais.info()
```

```python
#Padronizando nome das regiões:
for index, row in dados_finais.iterrows():
  if row['Region'] == 'East Asia':
    row['Region'] = 'Eastern Asia'
  elif row['Region'] == 'Southeast Asia':
    row['Region'] = 'Southeastern Asia'
  elif row['Region'] == 'North America and ANZ':
    row['Region'] = 'North America'
```

```python
#Padronizando região dos paises 'New Zealand' e 'Australia'
for index, row in dados_finais.iterrows():
    if row['Country'] == 'New Zealand' or row['Country'] == 'Australia':
      row['Region'] = 'Australia and New Zealand'
```

```python
#Nome de todas as regiões
dados_finais['Region'].unique()
```

```python
#Padronizando nome dos paises:
for index, row in dados_finais.iterrows():
    if row['Country'] == 'Hong Kong S.A.R. of China' or row['Country'] == 'Hong Kong' or row['Country'] == 'Hong Kong S.A.R., China':
      row['Country'] = 'Hong Kong S.A.R China'
    elif row['Country'] == 'Trinidad & Tobago':
      row['Country'] = 'Trinidad and Tobago'
    elif row['Country'] == 'Czechia':
      row['Country'] = 'Czech Republic'
    elif row['Country'] == 'North Macedonia':
      row['Country'] = 'Macedonia'
    elif row['Country'] == 'Cyprus' or row['Country'] == 'Northern Cyprus':
      row['Country'] = 'North Cyprus'
```

```python
#Removendo '*' dos nomes dos paises:
dados_finais['Country'] = dados_finais['Country'].apply(lambda x: x.replace('*', ''))
```

```python
#Nome de todos os paises:
dados_finais['Country'].unique()
```

```python
#Criando dicionários com as regiões faltantes:
pais_região = {'Taiwan Province of China':'Eastern Asia', 
  'Belize':'Latin America and Caribbean',
  'Hong Kong S.A.R China': 'Eastern Asia',
  'Somalia':'Middle East and North Africa',
  'Namibia':'Southern Asia',
  'South Sudan': 'Middle East and Northern Africa',
  'Trinidad and Tobago':'Latin America and Caribbean',
  'Northern Cyprus': 'Central and Eastern Europe',
  'North Macedonia': 'Central and Eastern Europe',
  'Gambia':'Sub-Saharan Africa',
  'Luxembourg':'Western Europe',
  'Czechia':'Central and Eastern Europe',
  'Guatemala':'Latin America and Caribbean',
  'Kuwait':'Middle East and North Africa',
  'Belarus':'Central and Eastern Europe',
  'Turkmenistan':'Central and Eastern Europe',
  'North Cyprus':'Western Europe',
  'Libya' : 'Middle East and Northern Africa',
  'Azerbaijan':'Central and Eastern Europe',
  'Gambia':'Sub-Saharan Africa',
  'Liberia':'Sub-Saharan Africa',
  'Congo': 'Middle East and Northern Africa',
  'Niger': 'Sub-Saharan Africa',
  'Comoros': 'Sub-Saharan Africa',
  'Palestinian Territories':'Middle East and Northern Africa',
  'King of Eswatini' : 'Southern Africa',
  'Madagascar':'Sub-Saharan Africa',
  'Chad':'Sub-Saharan Africa',
  'Yemen':'Middle East and Northern Africa',
  'Mauritania':'Sub-Saharan Africa',
  'Lesotho':'Sub-Saharan Africa',
  'Botswana':'Sub-Saharan Africa',
  'Rwanda':'Sub-Saharan Africa',
  'Macedonia' : 'Central and Eastern Europe',
  'Trinidad and Tobago' : 'Latin America and Caribbean',
  'Czech Republic': 'Western Europe',
  'Eswatini, Kingdom of': 'Sub-Saharan Africa'}
```

```python
#Atribuindo as regiões faltantes conforme pais:
for index, row in dados_finais.iterrows():
  for chave, valor in pais_região.items():
    if row['Country'] == chave:
      row['Region'] = pais_região[chave]
```

```python
#Convertendo o tipo das colunas para flot64 (colunas que precisam ser numéricas)

dados_finais['Happiness Rank'] = dados_finais['Happiness Rank'].astype('int')

dados_finais['Happiness Score'] = dados_finais['Happiness Score'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Happiness Score'] = dados_finais['Happiness Score'].astype('float64')

dados_finais['Economy (GDP per Capita)'] = dados_finais['Economy (GDP per Capita)'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Economy (GDP per Capita)'] = dados_finais['Economy (GDP per Capita)'].astype('float64')

dados_finais['Family (Social Support)'] = dados_finais['Family (Social Support)'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Family (Social Support)'] = dados_finais['Family (Social Support)'].astype('float64')

dados_finais['Health (Life Expectancy)'] = dados_finais['Health (Life Expectancy)'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Health (Life Expectancy)'] = dados_finais['Health (Life Expectancy)'].astype('float64')

dados_finais['Freedom'] = dados_finais['Freedom'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Freedom'] = dados_finais['Freedom'].astype('float64')

dados_finais['Trust (Government Corruption)'] = dados_finais['Trust (Government Corruption)'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Trust (Government Corruption)'] = dados_finais['Trust (Government Corruption)'].astype('float64')

dados_finais['Generosity'] = dados_finais['Generosity'].apply(lambda x: str(x).replace(",", '.'))
dados_finais['Generosity'] = dados_finais['Generosity'].astype('float64')
```
